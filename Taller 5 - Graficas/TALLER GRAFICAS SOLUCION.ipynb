```python
import sqlite3
import numpy as np
import matplotlib as plt
```

```python
class CovidRepository:
    table = "COVID_COL"
    db_url = "C:\\Users\\user\\Documents\\MAESTRIA\\PROGRAMACION\\Taller 5 - Graficas\\covid19.db"

    def __init__(self):
        self.__connection = sqlite3.connect(self.db_url)

    def get_casos_confirmados (self):
        try :
            query = f"""
                    SELECT 
                    	t.date, 
                    	SUM(t.confirmed)
                	FROM timeseries t  
                	WHERE t.id IN (
            			SELECT l.id
            			FROM location l
            			WHERE l.iso_alpha_2 = "CO")
                	GROUP BY t.date 
                	ORDER BY t.date ASC 
                    """
            cursor = self.__connection.cursor()
            cursor.execute(query)
            return cursor.fetchall()
        except sqlite3.Error as e:
            print("error:", e)
            raise e
```

```python
COVID_COL_repository = CovidRepository()
```
