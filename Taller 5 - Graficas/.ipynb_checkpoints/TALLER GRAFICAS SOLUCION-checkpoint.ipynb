```python
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class CovidRepository:
    table = "COVID_COL"
    db_url = "C:\\Users\\user\\Documents\\MAESTRIA\\PROGRAMACION\\GIT\\MGAD-SERGIO-ARBOLEDA-PROGRAMACION\\Taller 5 - Graficas\\covid19.db"

    def __init__(self):
        self.__connection = sqlite3.connect(self.db_url)

    def get_casos_confirmados (self):
        try :
            query = f"""
                    SELECT 
                    	t.date, 
                    	SUM(t.confirmed)
                	FROM timeseries t  
                	WHERE t.id IN (
            			SELECT l.id
            			FROM location l
            			WHERE l.iso_alpha_2 = "CO")
                	GROUP BY t.date 
                	ORDER BY t.date ASC 
                    """
            cursor = self.__connection.cursor()
            cursor.execute(query)
            return cursor.fetchall()
        except sqlite3.Error as e:
            print("error:", e)
            raise e

    def get_muertes (self):
        try :
            query = f"""
                    SELECT 
                    	t.date, 
                    	SUM(t.deaths)
                	FROM timeseries t  
                	WHERE t.id IN (
            			SELECT l.id
            			FROM location l
            			WHERE l.iso_alpha_2 = "CO")
                	GROUP BY t.date 
                	ORDER BY t.date ASC 
                    """
            cursor = self.__connection.cursor()
            cursor.execute(query)
            return cursor.fetchall()
        except sqlite3.Error as e:
            print("error:", e)
            raise e

    def get_recuperados (self):
        try :
            query = f"""
                    SELECT 
                    	t.date, 
                    	SUM(t.recovered)
                	FROM timeseries t  
                	WHERE t.id IN (
            			SELECT l.id
            			FROM location l
            			WHERE l.iso_alpha_2 = "CO")
                	GROUP BY t.date 
                	ORDER BY t.date ASC 
                    """
            cursor = self.__connection.cursor()
            cursor.execute(query)
            return cursor.fetchall()
        except sqlite3.Error as e:
            print("error:", e)
            raise e
```

```python
COVID_COL_repository = CovidRepository()
```

```python
print(COVID_COL_repository.get_casos_confirmados()[:5]) #la salida es una lista en tuplas

#print(y[:5])
```

```python
datos = COVID_COL_repository.get_casos_confirmados() #limpio los datos

x = []
y = []

for i in range(len(datos)):
    x.append(i)
   
    #y.append(datos[i][1])
    y.append(0 if datos[i][1] is None else datos[i][1])
```

```python
print(y[:5])
#print(x[:5])#imprimir los 5 primeros valores de la lista 
```

```python
plt.plot(x,y,label="#CASOS CONFIRMADOS")
plt.legend(loc ="upper left")

plt.suptitle("CASOS CONFIRMADOS COVID 19")
plt.ylabel("NUMERO DE CASOS")
plt.xlabel("PERIODOS")
plt.ticklabel_format(style='plain', axis='y') #Arroja los valores en y normales 

plt.show()
```
